```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/capitadata.csv')  # Replace with your file path

# Inspect the data
print(data.head())

# Rename columns for easier access
data.rename(columns={
    "Entity": "Country",
    "Annual CO₂ emissions (per capita)": "Emissions_per_capita"
}, inplace=True)

# Handle missing values (if any)
data.dropna(inplace=True)

# Check for duplicates
data.drop_duplicates(inplace=True)
```

# Global Average Per Capita CO₂ Emissions Over Time

```python
# Calculate the global average per capita emissions by year
global_avg = data.groupby("Year")["Emissions_per_capita"].mean()

# Plot the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(global_avg.index, global_avg.values, marker='o', color='b', label='Global Average')
plt.title("Global Average Per Capita CO₂ Emissions Over Time")
plt.xlabel("Year")
plt.ylabel("Emissions (per capita)")
plt.grid()
plt.legend()
plt.show()
```


**Observation:**

The global average per capita CO₂ emissions have steadily increased from the 19th century to the mid-20th century, with a significant rise during the industrial revolution.
After peaking in the 1960s–1980s, emissions have shown fluctuations but no clear decline, highlighting the ongoing reliance on fossil fuels.

**Implication:**

The rapid increase during the industrial era correlates with industrialization and urbanization. Despite advancements in renewable energy, global emissions remain a pressing issue.

# Top 5 Countries by Per Capita CO₂ Emissions (2023)

```python
# Filter data for the latest year
latest_year = data["Year"].max()
latest_data = data[data["Year"] == latest_year]

# Find the top 5 countries
top_5_countries = latest_data.nlargest(5, "Emissions_per_capita")

# Plot
plt.figure(figsize=(8, 5))
plt.bar(top_5_countries["Country"], top_5_countries["Emissions_per_capita"], color='green')
plt.title(f"Top 5 Countries by Per Capita CO₂ Emissions ({latest_year})")
plt.xlabel("Country")
plt.ylabel("Emissions (per capita)")
plt.xticks(rotation=45)
plt.show()
```

**Observation:**

Qatar has the highest per capita CO₂ emissions in 2023, followed by Brunei, Bahrain, Trinidad and Tobago, and Saudi Arabia.
These countries share common factors: smaller populations, high dependency on oil and gas industries, and affluent lifestyles.

**Implication:**

The high per capita emissions in these countries emphasize the environmental challenges of wealth combined with fossil fuel-driven economies.

# Per Capita CO₂ Emissions Trend: Afghanistan

```python
# Filter data for Afghanistan
afghanistan_data = data[data["Country"] == "Afghanistan"]

# Plot the trend
plt.figure(figsize=(10, 6))
plt.plot(afghanistan_data["Year"], afghanistan_data["Emissions_per_capita"], marker='o', color='orange')
plt.title("Per Capita CO₂ Emissions Trend: Afghanistan")
plt.xlabel("Year")
plt.ylabel("Emissions (per capita)")
plt.grid()
plt.show()
```

**Observation:**

Afghanistan’s per capita CO₂ emissions increased gradually from the 1950s to 1980, reflecting industrial and economic developments.
There’s a noticeable dip in emissions around the 1990s–2000s, potentially due to political instability and reduced industrial activity.
Emissions show a modest rise after 2010, indicating slow recovery in economic activities.

**Implication:**

Political and economic stability play a significant role in a country's emissions trends. Afghanistan's low emissions per capita highlight its limited contribution to global CO₂ levels despite its challenges.

Additional Insights

Countries with high per capita emissions often have higher energy consumption for industrial or domestic purposes.

Developing nations like Afghanistan and many African countries have significantly lower emissions per capita, showing disparities in global emissions responsibilities.
